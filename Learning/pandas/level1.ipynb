```python
import pandas as pd


users = pd.DataFrame([
    [1, 'Alice', 'female', '23', 'China ', '2023-01-15'],
    [2, 'Bob', 'male', '31', 'USA', '2023-03-20'],
    [3, 'Cathy', 'female', None, 'china', '2023-02-11'],
    [4, 'David', 'male', '27', 'UK', '2023-05-09'],
    [5, 'Eva', 'female', '35', 'USA ', '2023-04-01'],
    [6, 'Frank', 'male', '19', None, '2023-06-18'],
], columns=['user_id', 'name', 'gender', 'age', 'country', 'signup_date'])


products = pd.DataFrame([
    [101, 'Keyboard', 'Electronics', '199'],
    [102, 'Mouse', 'Electronics', '99'],
    [103, 'Monitor', 'Electronics', '1299'],
    [104, 'Desk', 'Furniture', '899'],
    [105, 'Chair', 'Furniture', '499'],
    [106, 'Notebook', 'Stationery', '15'],
], columns=['product_id', 'product_name', 'category', 'price'])


orders = pd.DataFrame([
    [1001, 1, 101, '2', '2024-01-03 10:15', 'paid', '0'],
    [1002, 2, 103, '1', '2024-01-05 14:20', 'paid', '100'],
    [1003, 1, 106, '5', '2024-01-08 09:00', 'paid', None],
    [1004, 3, 102, '3', '2024-01-08 21:10', 'cancelled', '0'],
    [1005, 4, 104, '1', '2024-02-01 11:30', 'paid', '50'],
    [1006, 5, 105, '2', '2024-02-03 16:45', 'paid', '0'],
    [1007, 2, 101, '1', '2024-02-10 08:00', 'refunded', '0'],
    [1008, 6, 106, '10', '2024-02-12 19:20', 'paid', '5'],
    [1009, 1, 103, '1', '2024-02-15 13:00', 'paid', '200'],
    [1010, 3, 105, '1', '2024-03-01 10:10', 'paid', None],
    [1011, 5, 101, '2', '2024-03-03 12:00', 'paid', '20'],
    [1012, 2, 104, '1', '2024-03-05 18:30', 'paid', '0'],
    [1013, 2, 104, '1', '2024-03-05 18:30', 'paid', '0'],   # 故意重复
], columns=['order_id', 'user_id', 'product_id', 'quantity', 'order_time', 'status', 'discount'])
```

```python
orders['final_amount'] = products.set_index('product_id').loc[orders['product_id'], 'price'].reset_index(drop=True).astype(float) * orders['quantity'] - orders['discount']
orders
```

```python
df_all = orders
df_all['user_name'] = df_all['user_id'].map(users.set_index('user_id')['name'])
df_all['user_country'] = df_all['user_name'].map(users.set_index('name')['country'])
df_all['product_name'] = df_all['product_id'].map(products.set_index('product_id')['product_name'])
df_all['product_category'] = df_all['product_name'].map(products.set_index('product_name')['category'])
df_all['product_price'] = df_all['product_name'].map(products.set_index('product_name')['price'])
df_all['final_amount'] = df_all['quantity'].astype(float) * df_all['product_price'].astype(float)
df_all
```

```python
df_all = orders.merge(
    users[['user_id', 'name', 'country']],
    on='user_id',
    how='left'
).merge(
    products[['product_id', 'product_name', 'category', 'price']],
    on='product_id',
    how='left'
)
df_all['discount'] = df_all['discount'].fillna(0)
df_all = df_all.rename(columns={
    'name': 'user_name',
    'country': 'user_country',
    'category': 'product_category',
    'price': 'product_price'
})

df_all['final_amount'] = df_all['quantity'].astype(float) * df_all['product_price'].astype(float) - df_all['discount'].astype(float)

df_all
```
